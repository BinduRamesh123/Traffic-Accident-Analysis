```python
df['Traffic_Density'] = df.groupby('Road_Type')['Traffic_Density']\
                      .transform(lambda x: x.fillna(x.median()))
```

```python
df['Speed_Limit'] = df.groupby('Road_Type')['Speed_Limit']\
                      .transform(lambda x: x.fillna(x.median()))
```

```python
df['Driver_Experience'] = df.groupby('Driver_Age')['Driver_Experience']\
                            .transform(lambda x: x.fillna(x.median()))
```

```python
df['Driver_Age']=df['Driver_Age'].fillna(df['Driver_Age'].median())
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=3)
df["Number_of_Vehicles"]=imputer.fit_transform(df[["Number_of_Vehicles"]])
```

```python
df.head()
```

```python
df['Driver_Alcohol'] = df['Driver_Alcohol'].fillna(df['Driver_Alcohol'].mode()[0])
df
```

```python
df['Accident_Severity'] = df['Accident_Severity'].fillna(df['Accident_Severity'].mode()[0])
```

```python
df['Road_Condition'] = df.groupby('Weather')['Road_Condition'] \
                          .transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df['Vehicle_Type'] = df['Vehicle_Type'].fillna(df['Vehicle_Type'].mode()[0])
```

```python
df['Driver_Age'] = df['Driver_Age'].fillna(df['Driver_Age'].median())
```

```python
df['Age_Group'] = pd.cut(df['Driver_Age'],
                         bins=[18,25,40,60,80],
                         labels=['Young','Adult','Middle','Senior'])
#new column
```

```python
df['Driver_Experience'] = df.groupby('Age_Group')['Driver_Experience'] \
                             .transform(lambda x: x.fillna(x.median()))
```

```python
df['Road_Light_Condition'] = df.groupby('Time_of_Day')['Road_Light_Condition'] \
                                .transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df['Accident'] = df['Accident'].fillna(df['Accident'].mode()[0])
```

```python
df.tail()
```

```python
acc_percent = df['Accident'].value_counts(normalize=True) * 100
acc_percent.index = ['No Accident', 'Accident']
plt.bar(acc_percent.index, acc_percent.values,
        color=['green', 'red'])
plt.title("Accident Distribution (%)")
plt.ylabel("Percentage")
plt.ylim(0,100)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Driver_Alcohol', hue='Accident', data=df)
plt.title("Alcohol Consumption vs Accident")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Accident', y='Speed_Limit', data=df)
plt.title("Speed Limit vs Accident")
plt.show()
```

```python
plt.figure(figsize=(7,4))
df['Accident_Label'] = df['Accident'].map({0: 'No Accident', 1: 'Accident'})
sns.countplot(x='Weather', hue='Accident_Label', data=df)
plt.title("Weather vs Accident")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Driver_Age',
                y='Driver_Experience',
                hue='Accident',
                data=df)
plt.title("Age vs Experience vs Accident")
plt.show()
```
